### Models Integration With Openai, Google Gemini, Groq

```python
import os
from dotenv import load_dotenv
load_dotenv()

os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY","")
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY","")
os.environ["GOOGLE_API_KEY"] = os.getenv("GOOGLE_API_KEY","")   
```

```python
from langchain.chat_models import init_chat_model
model = init_chat_model("gpt-4.1")
response = model.invoke("What is the capital of France?")
response.content
```

## ChatOpenAI

```python
from langchain_openai import ChatOpenAI
model = ChatOpenAI(model="gpt-4.1") 
response = model.invoke("What is the capital of France?")
response.content
```

### Google Gemini Integration

```python
from langchain.chat_models import init_chat_model
model = init_chat_model("google_genai:gemini-2.5-flash-lite")
response = model.invoke("What is the capital of France?")
response.content
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
model = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")
response = model.invoke("What is the capital of France?")
response.content
```

### GROQ Model Integration

```python
from langchain.chat_models import init_chat_model
model = init_chat_model("groq:qwen/qwen3-32b")
response = model.invoke("What is the capital of France?")
response.content
```

```python
from langchain_groq import ChatGroq
model = ChatGroq(model="qwen/qwen3-32b")
response = model.invoke("What is the capital of France?")
response.content
```

### STREAMING AND BATCH

```python
model.invoke("write an essay on the importance of education in modern society in 500 words")
```

```python
for chunk in model.stream("write an essay on the importance of education in modern society in 500 words"):
    print(chunk.text, end="")
```

### BATCH
Batch is a collection of independant request to a model can significantly improve performance and reduce costs, as the processing can be done in parallel:

```python
response = model.batch([
    "What is the capital of France?",
    "What is the capital of Germany?",
    "What is the capital of Italy?"])

for res in response:
    print(res)
```

```python
response = model.batch([
    "What is the capital of France?",
    "What is the capital of Germany?",
    "What is the capital of Italy?"],
    config={
        'max_concurrency':5, # Limit the number of parallel requests to 5
    }
    )
```
