### Messages

```python
from langchain.chat_models import init_chat_model
model = init_chat_model("groq:qwen/qwen3-32b")
response = model.invoke("What is the capital of France?")
response
```

### Text Prompts
Text prompts are strings - ideal for straightforward generation tasks where you dont need to retain the conversation history

```python
model.invoke("What is the capital of France?")
```

### Message Prompts
Alternatively you can pass in a list of messages to the model by providing a list of message objects

Message Type:
1. System Message
2. Human Message
3. AI Message
4. Tool Message


```python
from langchain.messages import SystemMessage, HumanMessage, AIMessage

messages = [
    SystemMessage("You are a helpful assistant."),
    HumanMessage("What is the capital of France?")
]

response = model.invoke(messages)
response.content
```

```python
msg = HumanMessage(
    content="What is the capital of France?",
    name="user1",
    id="msg1"
    
)
msg
```

```python
response = model.invoke([msg])
response
```

```python
from langchain.messages import SystemMessage, HumanMessage, AIMessage
ai_msg = AIMessage("I would be happy to help you with that.")
messages = [
    SystemMessage("You are a helpful assistant."),
    HumanMessage("can you help me",name="user1",id="msg1"),
    ai_msg,
    HumanMessage("Great! what is 2+2?",name="user1",id="msg2")
    ]

response = model.invoke(messages)
response
```

```python
from langchain.messages import AIMessage, ToolMessage
ai_msg = AIMessage(
    content = [],
    tool_calls = [{
        "name": "get_weather",
        "args": {"location": "New York"},
        "id": "tool_call_1"}
    ]
)

weather_result = "Sunny, 25¬∞C"
tool_msg = ToolMessage(
    content = weather_result,
    tool_call_id = "tool_call_1"
)

messages = [
    HumanMessage("What is the weather like in New York?", name="user1", id="msg1"),
    ai_msg,
    tool_msg
]

response = model.invoke(messages)
response
```
